```python
def plot_qty():
    plt.figure(figsize=(15,6))

    dates=df_nike['date']

    plt.plot(dates, df_nike["sold_qty"],marker="o",label="Nike")
    plt.plot(dates, df_adidas["sold_qty"],marker="o",label="Ad")

    plt.xlabel("Date")
    plt.ylabel("Total Qty Sold")
    plt.title("Daily Sales Qty in September")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid(True,which="both",linestyle="--",linewidth=0.5)
    plt.show()
```

```python
plot_qty()
```
