```python
df = pd.read_csv("weather_data.csv", parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index("day", inplace=True)
df
```

```python
zero_df = df.fillna(0)
zero_df
```

```python
df.temperature.mean()
```

```python
dict_df = df.fillna(
    {
        "temperature": df.temperature.mean(),
        "windspeed": df.windspeed.mean(),
        "event": "No Event",
    }
)
dict_df
```

```python
forward_fill_df = df.fillna(method="ffill")
forward_fill_df
```

```python
backward_fill_df = df.fillna(method="bfill")
backward_fill_df
```

```python
backward_fill_col_df = df.fillna(method="bfill",axis="columns")
backward_fill_col_df
```

```python
forward_fill_limit_df = df.fillna(method="ffill",limit=1)
forward_fill_limit_df
```
