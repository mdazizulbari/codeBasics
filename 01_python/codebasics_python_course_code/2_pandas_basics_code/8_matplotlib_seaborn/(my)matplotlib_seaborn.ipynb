```python
import pandas as pd
```

```python
df_sales = pd.DataFrame({
    'Quarter': ['Q1 2022', 'Q2 2022', 'Q3 2022', 'Q4 2022', 'Q1 2023'],
    'Fridge': [12.5, 11.0, 13.2, 16.4, 13.8],
    'Dishwasher': [8.0, 7.5, 9.0, 11.0, 9.5],
    'Washing Machine': [9.2, 8.5, 7.5, 7.6, 6.8]
})
df_sales.head()
```

```python
from matplotlib import pyplot as plt

# plt.figure()
plt.plot(df_sales["Quarter"],df_sales["Fridge"], color="red", label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"])

plt.title("Product Sales")
plt.ylabel("Revenue (mls $)")
plt.xlabel("Financial Quarter")
plt.legend()

plt.show()
```

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%", explode=(0.1,0.2,0),shadow=True,startangle=140)
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation=45)
plt.show()
```

```python
df_score = pd.DataFrame({
    'Name': ['Aarav', 'Michael', 'Bilal'],
    'Exam_Score': [45, 71, 59]
})

df_score.head(3)
```

```python
import seaborn as sns
```

```python
plt.hist(df_score["Exam_Score"])
```

```python
# sns.histplot(df_score["Exam_Score"],kde=True)
sns.histplot(data=df_score,x="Exam_Score",kde=True)

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```
